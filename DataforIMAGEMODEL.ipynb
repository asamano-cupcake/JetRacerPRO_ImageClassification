```python
import ipywidgets as widgets
from IPython.display import display
import cv2
import os

bike_dir = 'dataset/bike'
stop_dir = 'dataset/stop'
railroads_dir = 'dataset/railroads'
trafficlights_dir = 'dataset/trafficlights'
road_dir = 'dataset/road'

try:
    os.makedirs(bike_dir)
    os.makedirs(stop_dir)
    os.makedirs(railroads_dir)
    os.makedirs(trafficlights_dir)
    os.makedirs(road_dir)
except FileExistsError:
    print('Directories already exist')

button_layout = widgets.Layout(width='128px', height ='64px')
stop_button = widgets.Button(description='Add Stop', button_stype = 'danger', layout=button_layout)
bike_button = widgets.Button(description='Add Bike', button_stype = 'danger', layout =button_layout)
railroads_button = widgets.Button(description ='Add Railroad', button_stype = 'danger', layout=button_layout)
trafficlights_button = widgets.Button(description = 'Add TrafficLights', button_stype = 'danger', layout=button_layout)
road_button = widgets.Button(description = 'Add Road', button_stype = 'danger', layout=button_layout)

stop_count = widgets.IntText(layout=button_layout, value=len(os.listdir(stop_dir)))
bike_count = widgets.IntText(layout=button_layout, value=len(os.listdir(bike_dir)))
railroads_count = widgets.IntText(layout=button_layout, value=len(os.listdir(railroads_dir)))
trafficlights_count = widgets.IntText(layout=button_layout, value=len(os.listdir(trafficlights_dir)))
road_count = widgets.IntText(layout=button_layout, value=len(os.listdir(road_dir)))
```

```python
import traitlets
import ipywidgets.widgets as widgets
from IPython.display import display
from jetcam.utils import bgr8_to_jpeg
from jetcam.csi_camera import CSICamera

camera = CSICamera(width=224, height=224, capture_fps=65)
image_widget = widgets.Image(format='jpeg', width=224, height=224)
camera.running = True
camera_link = traitlets.dlink((camera, 'value'),(image_widget, 'value'), transform=bgr8_to_jpeg)
```

```python
from uuid import uuid1

count_stop = 0
count_bike = 0
count_railroads = 0
count_trafficlights = 0
count_road = 0

def save_snapshot(directory):
    image_path = os.path.join(directory, str(uuid1()) + '.jpeg')
    cv2.imwrite(image_path, camera.value.copy())

def save_stop():
    global stop_dir, stop_count
    save_snapshot(stop_dir)
    stop_count.value = len(os.listdir(stop_dir))
def save_bike():
    global bike_dir, bike_count
    save_snapshot(bike_dir)
    bike_count.value = len(os.listdir(bike_dir))

def save_railroads():
    global railroads_dir, railroads_count
    save_snapshot(railroads_dir)
    railroads_count.value = len(os.listdir(railroads_dir))

def save_trafficlights():
    global trafficlights_dir, trafficlights_count
    save_snapshot(trafficlights_dir)
    trafficlights_count.value = len(os.listdir(trafficlights_dir))
    
def save_road():
    global road_dir, road_count
    save_snapshot(road_dir)
    road_count.value = len(os.listdir(road_dir))

stop_button.on_click(lambda x: save_stop())
bike_button.on_click(lambda x: save_bike())
railroads_button.on_click(lambda x: save_railroads())
trafficlights_button.on_click(lambda x: save_trafficlights())
road_button.on_click(lambda x: save_road())
```

```python
display(image_widget)
display(widgets.HBox([stop_count, stop_button]))
display(widgets.HBox([bike_count, bike_button]))
display(widgets.HBox([railroads_count, railroads_button]))
display(widgets.HBox([trafficlights_count, trafficlights_button]))
display(widgets.HBox([road_count, road_button]))
```
